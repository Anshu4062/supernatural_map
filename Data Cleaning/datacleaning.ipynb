```python
# dependencies
import pandas as pd
from datetime import datetime as dt
```

```python
bf_loc = pd.read_csv("C:/Users/ginav/Desktop/Analysis Projects/Project 3/supernatural_map/Data Cleaning/resources/timothyrenner-bfro-sightings-data/bfro_locations.csv")
bfLocations = pd.DataFrame(bf_loc)
bfLocations.head()
```

```python
bf_geo = pd.read_csv("C:/Users/ginav/Desktop/Analysis Projects/Project 3/supernatural_map/Data Cleaning/resources/timothyrenner-bfro-sightings-data/bfro_reports_geocoded.csv")
bfGeocode = pd.DataFrame(bf_geo)
bfGeocode.head()
```

```python
# bfGeocode.columns
bfGeocode["number"] = bfGeocode["number"].astype(int)
bfGeocode.info()
```

```python
# bfLocations.info()
bfData = bfLocations.merge(bfGeocode, how='inner')
bfData.head()
```

```python
bfData = bfData.drop(columns=['uv_index', 'precip_intensity', 'precip_type', 'pressure', 'precip_probability', 'temperature_low', 'temperature_high', 'temperature_mid', 
                            'dew_point', 'humidity', 'cloud_cover', 'conditions', 'uv_index', 'visibility', 'wind_bearing','wind_speed', 'county'])
bfData.head()
```

```python
bfData = bfData.drop(columns=['season', 'moon_phase', 'geohash'])
bfData.head()
```

```python
bfData['timestamp'] = pd.to_datetime(bfData['timestamp'])
bfData.info()
```

```python
bfData = bfData.drop(columns=['date','classification'])
bfData.head()
```

```python
bfData = bfData.dropna()
bfData
```

```python
bfData.to_csv("C:/Users/ginav/Desktop/Analysis Projects/Project 3/supernatural_map/Data Cleaning/resources/bigfoot_data.csv")
```

```python
ufo_rep = pd.read_csv("C:/Users/ginav/Desktop/Analysis Projects/Project 3/supernatural_map/Data Cleaning/resources/timothyrenner-ufo-sightings/nuforc_reports.csv")
ufoReports = pd.DataFrame(ufo_rep)
ufoReports.head()
```

```python
ufoReports = ufoReports.drop(columns=['report_link', 'posted', 'city', 'stats', 'summary'])
ufoReports.head()
```

```python
ufoReports["date_time"] = pd.to_datetime(ufoReports["date_time"])
ufoReports.info()
```

```python
ufoReports = ufoReports.rename(columns={'date_time': 'datetime', 'city_latitude': 'latitude', 'city_longitude': 'longitude', 'text': 'description'})

ufoReports.head()
```

```python
ufo_clean = ufoReports.dropna()
ufo_clean
```

```python
ufo_sorted = ufo_clean.sort_values('datetime')
ufo_sorted = ufo_sorted.reset_index(drop=True)
ufo_sorted
```

```python
ufo_sliced = ufo_sorted.loc[ufo_sorted['datetime'] >= '2010']
ufo_sliced  
```
